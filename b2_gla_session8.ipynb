```python
# RNN ARCHITECTURE
# this will repressention of how hidden state gets updated
# This is not real RNN it's just readable simulisation
# It shows the main idea how the sequence reacts step by step
sentence = "The movie was not good"
tokens = sentence.split()

memory = ""

for token in tokens:
  memory = memory + " " + token
  print("Current token:", token)
  print("Memory so far:", memory.strip())
  print("-------")
```

```python
# proper RNN
# simplest version of RNN
# this how hidden state capture summary
sentence = "The movie was not good"
tokens = sentence.split()

hidden_state = {
    "topic": None,
    "negation_active": False,
    "sentiment_word": None
}

for token in tokens:
  if token in ["movie", "food", "product"]:
    hidden_state["topic"] = token

    # negation structure
  if token == "not":
    hidden_state["negation_active"] = True

  if token in ["good", "bad", "great", "terrible"]:
    hidden_state["sentiment_word"] = token

  print("Token:", token)
  print("Hidden state:", hidden_state)
  print("-----")








```

```python
# more simplify real life RNN architecture ( binary classification)
# tensorflow and embedding projector
import tensorflow as tf
from tensorflow.keras import layers

vocab_size = 1000
embedding_dim = 16 # 16 main paramters
sequence_length = 8

simple_rnn_model = tf.keras.Sequential([
    layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=sequence_length), # convert tokenid into values
    layers.SimpleRNN(32),
    layers.Dense(1, activation="sigmoid") # bcz output ( range  0 and 1 ) # it converts hidden state into prediction -- close to 1 intense +ve , and close to 0 _ve
])

simple_rnn_model.summary()
```
